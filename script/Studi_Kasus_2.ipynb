```python
# 1. Import library
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 2. Baca data
df = pd.read_csv('HASIL Query task 2.csv')

# 3. Konversi kolom 'encryption_level' ke angka
# Misalnya: 'High' = 3, 'Medium' = 2, 'Low' = 1
encryption_map = {'Low': 1, 'Medium': 2, 'High': 3}
df['encryption_level_num'] = df['encryption_level'].map(encryption_map)

# 4. Ambil data untuk visualisasi
x = df['amount_idr']
y = df['co2_sequestration_tonnes']
z = df['encryption_level_num']

# 5. Buat plot 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# 6. Scatter plot
scatter = ax.scatter(x, y, z, c=z, cmap='viridis', marker='o', edgecolor='k', s=60)

# 7. Label sumbu
ax.set_xlabel('Jumlah Dana (IDR)')
ax.set_ylabel('Penyerapan CO2 (ton)')
ax.set_zlabel('Tingkat Enkripsi (High=3)')

# 8. Judul
ax.set_title('3D Scatter: Jumlah Dana vs CO2 vs Enkripsi')

# 9. Warna berdasarkan tingkat enkripsi
cbar = plt.colorbar(scatter, ax=ax, shrink=0.5, pad=0.1)
cbar.set_label('Tingkat Enkripsi')

# 10. Tampilkan
plt.show()
```
